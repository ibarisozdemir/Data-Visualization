```python
# DataFrame oluşturma

boy = np.random.randint(150,190,(10))
kilo = np.random.randint(50,120,(10))
ogr_no = np.random.randint(2100,2200,(10))

my_dict = {"ogrenci_no":ogr_no,"ogr_boy":boy,"ogr_kilo":kilo}

my_df = pd.DataFrame.from_dict(my_dict)
my_df
```

```python
vize_not = np.random.randint(0,100,(10))
final_not = np.random.randint(0,100,(10))

my_df["ogr_vize"] = vize_not
my_df["ogr_final"] = final_not
my_df
```

```python
my_df["ogr_final"].values
```

```python
#ogr_gecti isimli bir key değeri ekleyin
# bu key değerine ait değerler, ogr_vize sütunundaki değerlerin
# yüzde kırkı ile ogr_final sütunundaki değerlerin yüzde altmışının
# toplamı olacak
```

```python
son_not = (my_df["ogr_vize"].values*0.4) + (my_df["ogr_final"].values*0.6)
 
my_df["ogr_gecti"] = son_not
```

```python
my_df["Durum"] = my_df["ogr_gecti"]>50
my_df
```

```python
# voleybol isminde bir key oluşturun
# eğer ogr_boy 175'ten büyükse ve ogr_kilo 70'ten büyükse
# Oynayabilir yazsın değilse oynayamaz yazsın


deneme = np.where((my_df["ogr_boy"]>175) & (my_df["ogr_kilo"]>75))
my_df.loc[deneme]
```
